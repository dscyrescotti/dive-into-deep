```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def train(x, y):
    m, b = 0, 0
    learning_rate = 0.001
    n = len(x)
    for i in range(1000):
        predict = m * x + b
        cost = (1/n) * sum([diff**2 for diff in (y-predict)])
        delta_m = -(2 / n) * sum(x * (y-predict))
        delta_b = -(2 / n) * sum(y-predict)
        m -= learning_rate * delta_m
        b -= learning_rate * delta_b
        print("m {}, b {}, c {}, i {}".format(m, b, cost, i))
```

```python
train(x, y)
```
